# Gradient Descent with Multiple Parameters

### Introduction

In past lessons, we saw how we can train our hypothesis function, that is, we saw how to discover the parameters of our hypothesis function that minimize our cost function.  We did this beginning with a random value for our parameter $w_1$, and then changing our parameter by an amount proportional to the value of the slope of our cost curve at a given point.  We also moved our weight in the opposite direction of the slope.  

For example, we saw that with given a neuron with a hypothesis function of the form: 

$$h(x) = \sigma(w_1x)$$

> Where $x$ represented the average cell area on a slide. 

We would start with a random value for $w_1$ and repeatedly update $w_1$ according to the formula the formula:

$w_1 = w_1 - learning\_rate*slope\_at(w_1)$ 

> Where `slope_at` is the slope of the cost curve at $w_1$.

<img src="./single-param-descent.png">

Now that are familiar with the derivative, we can be a little more formal.  If we define our function cost function as a function $J$ then instead of saying our procedure is to update $w_1$ by:

$w_1 = w_1 - learning\_rate*slope\_at(w_1)$ 

We express this as:

$w_1 = w_1 - learning\_rate*J'(w_1)$ 

Now that we have a sense of how to train a hypothesis function with a single parameter, let's update our procedure so that we can train a hypothesis function with multiple parameters.

### Training with multiple parameters

The first thing is to update our hypothesis function.  Over the past few lessons we've defined our hypothesis function to have just a single parameter.

$$g(x) = \sigma(w_1 x)$$

> Where $x$ is the average cell area of an observation.

Let's update our function so that now in addition to having a weight associated with cell area, it also has a bias term.

$$g(x) = \sigma(w_1x + b)$$

How do we find the value of this bias term?  Well the bias term is just another parameter in our hypothesis function that we have use training to learn.  So once again, we'll proceed with gradient descent, just this time we'll update our procedure for two parameters.  Remember that with gradient descent:

1. We'll begin with random values for our parameter  
2. We'll repeatedly update our parameter by calculating the slope of the cost curve at that point.  
* By doing so, we approach the minimum.  

But think about what our cost curve looks like with two parameters.  As we update either our weight or bias term, our hypothesis function will do better or worse at predicting the training data, and our cost curve will reflect that.

Yes, this is more complicated than our one dimensional cost curve.  But our approach of is the same.  We'll start at random values for our parameters, this time $w_1$ and $b$, and we'll use the slope of the cost curve at those parameters to approach our minimum.

### Descending along the cost curve

Ok, so we know that we use the slope to descend along the cost curve and thus discover the parameters that minimimize our cost.  But *how exactly* do we do this?  

The answer is that we try to maximize our descent with each step that we take.  And to do that, we move in proportion to the steepness in each direction.  So if our cost curve is three times steeper in the $x$ direction than the $y$ direction, we should step three times more in the $x$ direction than we do in the $y$ direction -- as that is the direction of greatest decrease.

This technique is called gradient descent.

We'll get into a more mathematical discussion of gradient descent in the next section, but for now, let's keep our discussion at the conceptual level.

### Descending a hill

To use a classic analogy, with gradient descent, the way that we find the minimum in our cost curve, is the same way we would efficiently walk down a hill.

<img src="./downhill-walk.png" width="40%">

But remember that we want to do this without taking the time to map out the entire cost curve.  So in taking that next step, we are essentially doing so blindfolded.  So if we were descending blindfolded, how would we know which direction to walk?

We lean to the right to see the steepness in one direction, and then lean forward to see the steepness in the other direction.  And then we walk in proportion to the steepness of each axis.  This tells us the how much the curve descends in each direction, and using that we can move in the direction of greatest decrease.

Now that we understand it, we can discuss what it means mathematically.  Doing involves -- this leaning in each direction -- involves an introduction to partial derivatives.

### Summary

In this lesson we learned how to use gradient descent to approach the minimum of a multi-dimensional cost curve.  With gradient descent, our technique is to take repeated steps in the direction of the greatest decrease.  Because we want to avoid calculating the entire cost curve, we find the direction of greatest decrease by finding the slope in one direction, and the other direction, and then move in a direction proportional to these slopes.
