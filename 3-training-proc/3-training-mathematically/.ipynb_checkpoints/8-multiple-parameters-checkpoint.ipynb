# Gradient Descent with Multiple Parameters

### Introduction

In past lessons, we saw how we can train our function, that is discover parameters that minimize a cost function, by looking at the slope of the cost curve at a given point.  For example, we saw that with given a sigmoid neuron of the form: 

$$g(x) = \sigma(w_1 * cell\_area)$$

We could find the parameter $w_1$ that best classified our cells as cancerous or not cancerous.

We did so by starting with a random value for $w_1$ and then updating $w_1$ with the formula:

$w_1 = w_1 - learning\_rate*slope\_at(w_1)$ 

Where `slope_at` is the slope of the cost curve at $w_1$.

In the past lessons, we got a sense of how we can find the slope of cost curve at a given point - by using the derivative.  In future lessons, we will apply our knowledge of the derivative to finding the slope at a given point. ### Training with multiple parameters

We previously simplified our problem, by using the hypothesis function.

$$g(x) = \sigma(w_1 * cell\_area)$$

But our linear component (or weighted input as it's also called) will not normally have just parameter.  Even a relatively simple linear component will have a bias term.  Let's update our function accordingly.

$$g(x) = \sigma(w_1 * cell\_area + w_0)$$

Our bias term is just a number -- like the number 10 or -12.  But what is that term?  Well to discover that term, we again want to essentially, use the same technique as previously but modify it for two parameters.  So now we will:

Begin with a random values for $w_1$ and $w_2$ and update them by using the slope.  Doing so will allow us to approach the minimum.  

But that with a hypothesis function with two different parameters, our cost curve will vary as these two parameters vary.  And thus our cost curve will be in two dimensions.  This is what it looks like.

### Descending along the cost curve

Ok, so we know that we use the slope to descend along the cost curve and thus discover the parameters that minimimize our cost.  But how exactly do we do this?  So far we have been purposely vague.  

The answer is that we try to maximize our descent with each step that we take.  And to do that, we move in proportion to the steepness in each direction.  So if our curve is three times steeper in the $x$ direction as the $y$ direction, we will step three times more in the $x$ direction than we do in the $y$ direction.

This technique is called gradient descent.

We'll get into a more mathematical discussion of gradient descent in the next section, but for now, let's keep our discussino at the conceptual level.

### Descending a hill

The way we find the minimum of our cost curve, is the same way we could efficiently walk down a hill.  

<img src="./downhill-walk.png" width="40%">

Remember, the goal is to descend as much as possible with each equally sized step.  Now remember that we want to do this without taking the time to map out the entire cost curve, so in taking that next step, we are essentially doing so blindfolded.  So how do we know which direction to walk?

Well we lean to the left and right, and then lean forward and backwards.  And we walk in proportion to the steepness of each axis.  This our gradient descent technique.

Now that we understand it, we can discuss what it means mathematically.  Doing involves an introduction to partial derivatives.
