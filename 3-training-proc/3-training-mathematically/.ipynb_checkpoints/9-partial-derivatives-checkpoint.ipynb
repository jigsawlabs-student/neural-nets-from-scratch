```python
def f(x, y):
    return 3*x*(y**2)
```

```python
from partial import x_vals, y_vals

z_vals =[f(w_1, w_2) 
         for (w_1, w_2) in list(zip(x_vals, y_vals))]
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=[
    go.Mesh3d(
        x=x_vals,
        y=y_vals,
        z=z_vals,
        colorbar_title='z',
        name='y',
        showscale=True
    )
])
fig
```

```python
def f(x, y):
    return 3*x*y**2
```

```python
vals = list(range(-10, 10))
scatter_at_six = go.Scatter(x=vals, y=[f(6,val) for val in vals], name = '6y^2')
scatter_at_four = go.Scatter(x=vals, y=[f(4,val) for val in vals], name = '4y^2')
scatter_at_two = go.Scatter(x=vals, y=[f(2,val) for val in vals], name = '2y^2')
scatter_at_one = go.Scatter(x=vals, y=[f(1,val) for val in vals], name = 'y^2')
go.Figure(data=[scatter_at_six, scatter_at_four, scatter_at_two])
```
