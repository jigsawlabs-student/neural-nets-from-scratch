# Machine learning Terms

### Introduction

In this lesson, it's time to learn about how machine learning occurs in practice.  We'll see that it begins gathering the data, and then moves onto training a machine learning model to arrive at a hypothesis function that predicts future outcomes.  Let's try to better understand what each of these steps mean.

### 1. Gather the data


Whenever we are faced with a machine learning algorithm, our first step is to gather our data.  For example, one of the most famous datasets in machine learning is the breast cancer cell dataset created in 1992 by Dr. Wolberg at the University of Wisconsin.

Dr. Wolberg had graduate students examine images of different cells, write down characteristics about them, and then specify whether the cells were cancerous or not.  

<img src="https://storage.cloud.google.com/curriculum-assets/nn-from-scratch/breast-cancer-study.jpeg" width="40%">

The goal was to learn characteristics of cells that could predict the presence of cancer.  For example this is a portion of the data he collected.

| mean area        | mean concavity    | cancerous
| ------------- |:-------------:| :-------------:| 
|    458.7        |0.02383  | 1  | 
|    1509       |0.42640 | 0 | 
|    298.3        |0.06154 | 1 | 
|    530.2     | 0.20710 | 0 | 
|    612.2     | 0.81678 | 1 | 
|    420.8     | 0.4578 | 0 | 

The data above represents our **training data**.  Gathering our training data is the first step in machine learning.  

> Our **training data** is a set of previous examples or observations, almost like a series of experiments.  

Our training data includes: 

1. Potential inputs to what we are trying to predict, and 
2. The past outcomes associated with these different inputs

### 2. Defining Terms

1. Observations

Another word we'll use for training data is **observations**.  So the first row of our data is our first *observation*, where we observed mean area of $458.7$, mean concavity of $.0238$, and that the corresponding cell is cancerous - as indicated by a 1.

| mean area        | mean concavity    | cancerous
| ------------- |:-------------:| :-------------:| 
|    458.7        |0.02383  | 1  | 

2. Target

Our target is what we are trying to predict.  Here our target is `cancerous`.  Given information about a cell, our goal is to be able to predict whether or not that cell is cancerous.

3. Features 

The features are the *inputs*, or the factors involved in making our prediction.  So above, our features are `mean area` and `mean concavity`.  This is what we are currently considering when making our prediction.

### Viewing Our Data

Let's take a look at how we can view some of this data.  Our data is currently in a csv file.  We'll use pandas to load up the data.

> Press `shift + enter` on the cell below to load the data.

```python
import pandas as pd
url = "https://raw.githubusercontent.com/jigsawlabs-student/neural-nets-from-scratch/master/0-intro-ml/cell_data.csv"
df = pd.read_csv(url)
df[:2]
```

Ok next we can plot each of our features using the plotly library.

```python
import plotly.graph_objects as go
layout = go.Layout(xaxis = {'title': 'mean area'}, 
                   yaxis = {'title': 'mean concavity', 'range': [0, .5]}, 
                   title = 'Is it cancerous?')
scatter_data = go.Scatter(x=df['mean_area'], y=df['mean_concavity'], 
                     marker=dict(color=df['is_cancerous'], colorscale='Viridis'), 
                     mode='markers', hovertext=df['is_cancerous'])
go.Figure(scatter_data, layout = layout)
```

Let's take a moment to interpret the graph above.  The yellow markers represent images of cells where there cancer was present, and the purple markers represent images of cells that were benign.

Now from the plot above, it appears that by drawing a line diagonally through the data, we could predict whether cells are cancerous or not.  Now we won't be correct 100% of the time in our prediction, but this could still be useful in an ultimate diagnosis.

### Making predictions

Once we have our training data we can train a machine learning model to find a pattern in the data.  The goal of training is to arrive at a hypothesis function that will allow us to predict the outcomes of future observations. 

Let's look at the graph below to see what this means.

```python
values = list(range(550, 800, 50))
boundaries = [0.41661, 0.3011275, 0.18563, 0.0701, -0.045341]
decision_trace = go.Scatter(x=values, y=boundaries, mode = 'lines', name='decision boundary')
go.Figure(data = [scatter_data, decision_trace], layout = layout)
```

The red line above is an example of a hypothesis function for a machine learning model.

> A hypothesis function takes in features of observations as it's inputs and comes close to predicting the targets of the observations as it's output. 

So above, the model predicts that everything to the left of the line is cancerous and that everything to the right of the line is not cancerous.  

As we mentioned previously, our line does not perfectly classify cancerous cells from non-cancerous cells.  Every time we see a purple marker on the left side of our line, that is an instance of our model making an error.  But that's ok -- for right now, our point isn't to construct the perfect machine learning model but rather to get a sense of the components of machine learning.
