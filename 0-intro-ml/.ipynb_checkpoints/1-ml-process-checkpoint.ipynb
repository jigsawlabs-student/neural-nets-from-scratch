# Machine learning Terms

### Introduction

Let's forget about neural networks for right now, and just focus on better understanding our machine learning process in general.

We'll see that in machine learning, we begin by first gathering the data, and then train a machine learning model to arrive at a hypothesis function that predicts future outcomes.  



### 1. Gather the data


Whenever we are faced with a machine learning algorithm, our first step is to gather our data.  For example, one of the most famous datasets in machine learning is the breast cancer cell dataset created in 1992 by Dr. Wolberg at the University of Wisconsin.

Dr. Wolberg had graduate students examine images of different cells, write down characteristics about them, and then specify whether the cells were cancerous or not.  

<img src="https://storage.cloud.google.com/curriculum-assets/nn-from-scratch/breast-cancer-study.jpeg" width="40%">

The goal was to learn characteristics of cells that could predict the presence of cancer.  For example this is a portion of the data he collected.

| mean area        | mean concavity    | cancerous
| ------------- |:-------------:| :-------------:| 
|    458.7        |0.02383  | 1  | 
|    1509       |0.42640 | 0 | 
|    298.3        |0.06154 | 1 | 
|    530.2     | 0.20710 | 0 | 
|    612.2     | 0.81678 | 1 | 
|    420.8     | 0.4578 | 0 | 

The data above represents our **training data**.  Gathering our training data is the first step in machine learning.  

> Our **training data** is a set of previous examples or observations, almost like a series of experiments.  

Our training data includes: 

1. Potential inputs to what we are trying to predict, and 
2. The past outcomes associated with these different inputs

### 2. Defining Terms

1. Observations

Our training data consists of a set of **observations**.  The first row of our data is our first *observation*, where we observed mean area of $458.7$, mean concavity of $.0238$, and that the slide does have cancerous cells - as indicated by a 1.

| mean area        | mean concavity    | cancerous
| ------------- |:-------------:| :-------------:| 
|    458.7        |0.02383  | 1  | 

2. Target

Our target is what we are trying to predict.  Here our target is `cancerous`.  Given information about cells on a slide, our goal is to be able to predict whether or not those cells are cancerous.

3. Features 

The features are the *inputs*, or the factors involved in making our prediction.  So above, our features are `mean area` and `mean concavity`.  This is what we are currently considering when making our prediction.

### Viewing Our Data

Let's take a look at how we can view some of this data.  Our data is currently in a csv file located [here](https://raw.githubusercontent.com/jigsawlabs-student/neural-nets-from-scratch/master/0-intro-ml/cell_data.csv).  We'll use pandas to load up the data.

> Press `shift + enter` on the cell below to load the data.

```python
import pandas as pd
url = "https://raw.githubusercontent.com/jigsawlabs-student/neural-nets-from-scratch/master/0-intro-ml/cell_data.csv"
df = pd.read_csv(url)
df[:2]
```

Next we can plot each of our features using the plotly library.

> Press shift + return below.

```python
cancerous_df = df[df['is_cancerous'] == 0]
benign_df = df[df['is_cancerous'] == 1]
```

```python
import plotly.graph_objects as go
layout = go.Layout(xaxis = {'title': 'mean area'}, 
                   yaxis = {'title': 'mean concavity', 'range': [0, .5]}, 
                   title = 'Is it cancerous?')
scatter_cancerous = go.Scatter(x=cancerous_df['mean_area'], y=cancerous_df['mean_concavity'], 
                     marker=dict(colorscale='Viridis'), 
                     mode='markers', name = 'cancerous')
scatter_benign = go.Scatter(x=benign_df['mean_area'], y=benign_df['mean_concavity'], 
                     marker=dict(colorscale='Viridis'), 
                     mode='markers', name = 'benign')
go.Figure([scatter_cancerous, scatter_benign], layout = layout)
```

Let's take a moment to interpret the graph above.  The yellow markers represent images of cells where there cancer was present, and the purple markers represent images of cells that were benign.

### Making predictions

Now from the plot above, it appears that by drawing a line diagonally through the data, we could predict whether cells are cancerous or not.  Now we won't be correct 100% of the time in our prediction, but this could still be useful in an ultimate diagnosis.

```python
import plotly.graph_objects as go
values = list(range(550, 800, 50))
boundaries = [0.41661, 0.3011275, 0.18563, 0.0701, -0.045341]
decision_trace = go.Scatter(x=values, y=boundaries, mode = 'lines', name='decision boundary')
fig = go.Figure(data = [scatter_cancerous, scatter_benign, decision_trace], layout = layout)
fig
```

From that pattern, we arrive at a hypothesis function which above is our red line.

> A **hypothesis function** takes in features of observations as it's inputs and comes close to predicting the targets of the observations as it's output. 

* Training

How did we get to that red line?  Well you can imagine us trying lines with different angles, or moving our red line further to the right or left to predict if cells are cancerous or not.  The hypothesis function that makes predictions that match what we actually observed, is the one that we'll use going forward.

> This is called **training**.  In training, we try different versions of a hypothesis function until we find a hypothesis function that comes close to predicting the targets that observed in our training data.

So for example, above, the red line comes close to predicting that certain observations contain cancer and others do not.  And you can imagine that we trained our hypothesis function by 

1. Trying different versions of the line.  
2. Choosing the hypothesis function that best matches what was observed in the training data.  

So for this data, the red line at the angle and position above made predictions of cancerous or not that best matched what was observed in our training data.

> We'll elaborate on what it means to "best match" in future lessons.

```python
fig
```

* Predict

Once we have used training to arrive at a hypothesis function, we can use our hypothesis function to make predictions going forward.  For example, if we see have the features of a new slide of cells, we can use the hypothesis function to predict whether cells on the slide are cancerous or not.

Now our hypothesis function above isn't perfect.  But that's ok -- for right now, our point is simply to get a sense of the components of machine learning.

### Summary

In this lesson, we saw the steps involved in to make predictions with a machine learning model.  The first step is to gather the training data.  

| mean area        | mean concavity    | cancerous
| ------------- |:-------------:| :-------------:| 
|    458.7        |0.02383  | 1  | 

Each row of data is referred to as an observation.  Each observation has a target, which is what we are trying to predict.  In our example, above our target was whether or not there was the presence of cancer.  To predict our target, we consider different characteristics in each observation, called our features -- above `mean area` and `mean concavity`.

* Training 

Once we have our data, we **train** a machine learning to find a pattern in the data and arrive at a hypothesis function.  

<img src="./cancer-plot.png" width="40%">

In training we try different versions of our hypothesis function.  Above by changing the angle or position of our red line.  We choose the hypothesis function whose predictions best match what we observed in our training data.

* Predicting

Once we have found the hypothesis function that best matches our training data, we can then use that hypothesis function to make predictions of observations going forward.  So here, we could look at the mean area and concavity of a future slide of cells to predict if cancer was present or not.
